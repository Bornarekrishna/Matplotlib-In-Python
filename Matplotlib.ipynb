```python
# First create  some catogorical and numerical Data
Price1 = [13,38,28,23,40,36,20,21,19,27]
Price2 = [33,22,32,31,26,48,23,10,35,37]
Price3 = [34,49,16,32,24,20,41,12,15,20]
Years = [2014,2015,2016,2017,2018,2019,2020,2021,2022,2023]

# Colour(Hex Code) and Line(Width And Style) and Marker(Type and Size)
plt.plot(Years,Price1, color="blue" , linestyle="dashed" , linewidth=1 , marker="D" , markersize= 5, label="P1")
plt.plot(Years,Price2, color="red" , linestyle="dotted" , linewidth=2 , marker="D", markersize= 5, label="P2")
plt.plot(Years,Price3, color="green" , linestyle="dashdot" , linewidth=3 , marker="D" , markersize= 5, label="P3")
# Labels
plt.title("Price Comparison") # Main Title
plt.xlabel("Years") # X axis label
plt.ylabel("Price") # Y axis label
# Legend
plt.legend()
# Limiting Axes
plt.xlim(2010,2025) # x Axis Limit
plt.ylim(10,60) # y Axis Limit
# Grid Lines
plt.grid()
```

# **Bar Plot**

```python
# First create catogorical and numerical Data
People = [98,60,80,70,55]
Cars = ["Hyundai","Toyota","Suzuki","Jeep","Audi"]

# plot a graph by using plt.bar() function
plt.bar(Cars,People)
```

```python
# Create DataFrame
Sales = {"Cars":["Hyundai","Toyota","Suzuki","Jeep","Audi"],
         "2020":[95,115,90,85,70],
         "2021":[100,95,120,105,90],
         "2022":[90,100,150,125,110]}

Sales = pd.DataFrame(Sales)

# Create Horizontal Bar Chart
plt.barh(Sales["Cars"],Sales["2020"])
```

```python
# I am using Sales Dataframe
# First Convert your X Axis Data in Array,like this
Bar = np.arange(len(Sales["Cars"]))

# creates Multiple Bar charts
plt.bar(Bar - 0.2 ,Sales["2020"],color="r",width=0.3 ,label="2020")
plt.bar(Bar ,Sales["2021"],color="b",width=0.3 ,label="2021")
plt.bar(Bar + 0.2 ,Sales["2022"],color="k",width=0.3 ,label="2022")

# restore X axis
plt.xticks(Bar,Sales["Cars"])

# Labels
plt.title("Sales Comparison") # Main Title
plt.xlabel("Cars Company") # X axis label
plt.ylabel("Units") # Y axis label
# Legend
plt.legend()
```

```python
# I am using Sales Dataframe

# Plot bars in stack manner
plt.bar(Sales["Cars"],Sales["2020"],color="y",label="2020")
plt.bar(Sales["Cars"],Sales["2021"],bottom=Sales["2020"],color="g",label="2021")
plt.bar(Sales["Cars"],Sales["2022"],bottom=Sales["2020"] + Sales["2021"],color="k",label="2022")

# Labels
plt.title("Sales Comparison") # Main Title
plt.xlabel("Cars Company") # X axis label
plt.ylabel("Units") # Y axis label
# Legend
plt.legend()
```

# Pie Chart

```python
# First create catogorical and numerical Data
Subject = ["English","Hindi","Maths","Science","History + Geography"]
Marks = [81,55,91,65,77]

# plot a graph by using plt.pie() function
plt.pie(Marks ,labels = Subject)
plt.show() # Display the plot
```

```python
# Create A Data
Product = {"Languages":["Python","JAVA","Javascript"," C and C++","SQL","HTML"],
           "Popular":[750000,430000,550000,310000,200000,370000]}
# Create A Dataframe
PRODUCT = pd.DataFrame(Product)

# Create Pie Plot
plt.pie(PRODUCT["Popular"],labels=PRODUCT["Languages"],autopct='%0.1f%%')
# autopct='%0.1f%%' use for show percentage on Pie chart

# Set Title
plt.title('Pie Chart')
plt.show() # Display the plot
```

```python
# I am using PRODUCT DataFrame
# Create Pie Plot
plt.pie(PRODUCT["Popular"],labels=PRODUCT["Languages"],autopct='%0.1f%%',
        colors=["#4DFF00","#FF2700","#6E5CB4","#ADA3AC","#FFF300","#FF00EC"],
        explode=[0.1,0,0,0,0,0],shadow=True)
# autopct='%0.1f%%' use for show percentage on Pie chart

# Set Title
plt.title('Pie Chart')
plt.show() # Display the plot
```

```python
# I am using PRODUCT DataFrame
# Create Pie Plot
plt.pie(PRODUCT["Popular"],labels=PRODUCT["Languages"],
        colors=["#4DFF00","#FF2700","#6E5CB4","#ADA3AC","#FFF300","#FF00EC"],
        hatch=['0.', '+', '\|', 'XX', 'O', '.', '*'])
# autopct='%0.1f%%' use for show percentage on Pie chart

# Set Title
plt.title('Pie Chart')
plt.show() # Display the plot
```
